```python
%autoreload 2

test = pd.read_csv('data/test.csv')
train = pd.read_csv('data/train.csv')
```

```python
test.head()
```

```python
#baseline model Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split    
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

X = pd.get_dummies(train[features],dtype=int)
Y = train['Survived']

X_train , X_valid , Y_train , Y_valid = train_test_split(X,Y, test_size=0.2, random_state=62)

model = RandomForestClassifier(n_estimators=100, random_state=62)
model.fit(X_train, Y_train)
valid_preds = model.predict(X_valid)
valid_acc = accuracy_score(Y_valid, valid_preds)

print(f"valid_acc is {valid_acc}")

X_test = pd.get_dummies(test[features],dtype=int)
test_preds = model.predict(X_test)
baseline_submission =pd.DataFrame({'passengerId':test['PassengerId'], 'Survived':test_preds})
baseline_submission.to_csv("data/baseline_submission.csv", index=False)
baseline_submission.head()
```
